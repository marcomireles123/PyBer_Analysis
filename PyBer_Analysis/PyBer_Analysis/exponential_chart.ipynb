```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis = np.arange(0, 5, 0.1)
x_axis
```

```python
e_x = [np.exp(x) for x in x_axis]
e_x
```

```python
plt.plot(x_axis, e_x)
plt.show
```

```python
plt.plot(x_axis, e_x)

plt.xlabel('Time with matplotlib')
plt.ylabel('How cool matplotlib seems')

plt.show()
```
